```python
import pandas as pd
import numpy as np
import os
```

```python
climate_list = os.listdir('D:/abstract/climate change/2020')
```

```python
for file in climate_list:
    df = pd.read_excel(os.path.join('D:/abstract/climate change/2020', file))
    if file == climate_list[0]:
        climate_file = df
    else:
        climate_file = pd.concat([climate_file, df], ignore_index=True)
```

```python
climate_file
```

```python
climate_file.drop_duplicates(subset=['UT (Unique WOS ID)'], keep='first', inplace=True)
climate_file.reset_index(drop=True, inplace=True)
```

```python
climate_file
```

```python
climate_file.to_csv('./climate_second.csv', sep=',')
```
