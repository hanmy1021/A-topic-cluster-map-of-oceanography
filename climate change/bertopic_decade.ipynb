```python
import pandas as pd 
import numpy as np 
from bertopic import BERTopic 
import os
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer
import joblib
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_dir = 'D:/abstract/climate change/first_inter.csv'
df = pd.read_csv(data_dir, engine='python')

df_drop = df.dropna(subset='Abstract')
docs = df_drop['Abstract'].to_list()

print(len(docs))
vectorizer_model = CountVectorizer(stop_words="english")
cluster_model = KMeans(n_clusters=20)
model1 = BERTopic(hdbscan_model=cluster_model, vectorizer_model=vectorizer_model, verbose=True)

topics, probabilities = model1.fit_transform(docs)
df_drop['topic'] = topics
df_drop.to_csv('D:/abstract/climate change/first_inter_withtopic.csv', index=False)
```

```python
model1.visualize_barchart(top_n_topics=20)
```

```python
first_topic = model1.get_topics()
```
