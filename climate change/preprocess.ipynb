```python
import pandas as pd
import numpy as np
import os
```

```python
climate_list = os.listdir('D:/abstract/climate change/2012-2021')
```

```python
for file in climate_list:
    df = pd.read_excel(os.path.join('D:/abstract/climate change/2012-2021', file))
    if file == climate_list[0]:
        climate_file = df
    else:
        climate_file = pd.concat([climate_file, df], ignore_index=True)
```

```python
climate_file.drop_duplicates(subset=['UT (Unique WOS ID)'], keep='first', inplace=True)
climate_file.reset_index(drop=True, inplace=True)
```

```python
climate_file.to_csv('D:/abstract/climate change/third_decade.csv', sep=',')
```

```python
ocean_file=pd.read_csv('D:/abstract/wos_0531/bib_360446/dropna_abstract_334765.csv',engine='python')
```

```python
WOS_ocean=list(ocean_file['UT (Unique WOS ID)'])
```

```python
WOS_first=list(first['UT (Unique WOS ID)'])
WOS_second=list(second['UT (Unique WOS ID)'])
WOS_third=list(third['UT (Unique WOS ID)'])
```

```python
first_inter=set(WOS_first).intersection(WOS_ocean)
second_inter=set(WOS_second).intersection(WOS_ocean)
third_inter=set(WOS_third).intersection(WOS_ocean)
```

```python
list(first_inter)
```

```python
first_intersection = first[first['UT (Unique WOS ID)'].isin(list(first_inter))] # 2866
second_intersection = second[second['UT (Unique WOS ID)'].isin(list(second_inter))] # 11335
third_intersection = third[third['UT (Unique WOS ID)'].isin(list(third_inter))] # 38015
```

```python
first_intersection.to_csv('D:/abstract/climate change/first_inter.csv', sep=',')
second_intersection.to_csv('D:/abstract/climate change/second_inter.csv', sep=',')
third_intersection.to_csv('D:/abstract/climate change/third_inter.csv', sep=',')
```
