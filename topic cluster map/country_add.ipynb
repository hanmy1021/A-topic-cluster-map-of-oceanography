```python
import pandas as pd
import os
import re
```

```python
df = pd.read_csv('D:/abstract/wos_0531/bib_360446/country_statistic.csv', engine='python')
```

```python
df = df[:15]
```

```python
del df['dir']
```

```python
df
```

```python
for i in range(15):
    df.loc[i,'IP%'] = df.loc[i,'IP']/df.loc[i,'P']
    df.loc[i,'CP%'] = df.loc[i,'CP']/df.loc[i,'P']
```

```python
df
```

```python
df['C']=df['C'].astype(int)
df['IP']=df['IP'].astype(int)
df['CP']=df['CP'].astype(int)
df['FP']=df['FP'].astype(int)
df['MCP']=df['MCP'].astype(int)
```

```python
df['IP%']=df['IP%']*100
df['CP%']=df['CP%']*100
df['FP%']=df['FP%']*100
df['MCP%']=df['MCP%']*100
```

```python
df
```

```python
df['P%']=df['P%'].round(decimals=2)
df['CPP']=df['CPP'].round(decimals=2)
df['IP%']=df['IP%'].round(decimals=2)
df['ICPP']=df['ICPP'].round(decimals=2)
df['CP%']=df['CP%'].round(decimals=2)
df['CCPP']=df['CCPP'].round(decimals=2)
df['FP%']=df['FP%'].round(decimals=2)
df['MCP%']=df['MCP%'].round(decimals=2)
```

```python
df.loc[0,'MC']='CHINA'
df.columns=['Country','P','P%','C','CPP','IP','IP%','CPIP','CP','CP%','CPCP','FP','FP%','MC','MCP','MCP%']
```

```python
df
```

```python
df.to_csv('D:/abstract/wos_0531/bib_360446/country_statistic_15.csv',index=False)
```
