```python
data_dir = 'D:/abstract/wos_0531/bib_360446/country_number/'
country_list = os.listdir(data_dir)
```

```python
country_list
```

```python
df = pd.DataFrame(columns=['Country', 'P', 'P%', 'dir'])
i = 0
total_number = 0
for country_ in country_list:
    country = country_.split('_')[0]
    number = country_.split('_')[1].split('.')[0]
    df.loc[i, 'Country'] = country
    df.loc[i, 'P'] = int(number)
    df.loc[i, 'P%'] = int(number)/353659*100
    df.loc[i, 'dir'] = country_
    i = i+1
    total_number += int(number)
```

```python
total_number
```

```python
df_country = df.sort_values(by='P', ascending=False)
```

```python
df_country = df_country.reset_index(drop=True)
```

```python
df_country[:20]
```

```python
data_dir = 'D:/abstract/wos_0531/bib_360446/country_number/'
for i in range(150,151):
    country_dir = data_dir + df_country.loc[i, 'dir']
    country_df = pd.read_csv(country_dir, engine='python')
    df_country.loc[i, 'C'] = sum(country_df['Times Cited, WoS Core'])
    df_country.loc[i, 'CPP'] = sum(country_df['Times Cited, WoS Core'])/df_country.loc[i, 'P']
    for j in range(len(country_df)):
        if type(country_df.loc[j, 'Addresses']) != float:
            country_address = re.sub('\[.*?\]','',country_df.loc[j, 'Addresses'])
            #country_address = 'Chad'
            addresses = country_address.split(';')
            author_countries = []
            for a in addresses:
                country_name = a.split(',')[-1].strip()
                if 'USA' in country_name:
                    country_name = 'USA'
                author_countries.append(country_name)
            country_df.loc[j, 'First Country'] = author_countries[0]
            country_df.loc[j, 'Countries'] = ','.join(set(author_countries))
            country_df.loc[j, 'If Independent'] = (len(list(set(author_countries)))==1)
    country_df.to_csv('D:/abstract/wos_0531/bib_360446/country_number/top20/'+df_country.loc[i, 'dir'], index=False)
```

```python
data_dir = 'D:/abstract/wos_0531/bib_360446/country_number/top20/'

i = 0
country_dir = data_dir + df_country.loc[i, 'dir']
country_df = pd.read_csv(country_dir, engine='python')

IP = 0
CP = 0
FP = 0
IPP = 0
CPP = 0
for j in range(len(country_df)):
    if country_df.loc[j, 'First Country'] == 'USA':
        FP += 1
        if country_df.loc[j, 'If Independent'] == True:
            IP += 1
            IPP += int(country_df.loc[j, 'Times Cited, WoS Core'])
        else:
            CP += 1
            CPP += int(country_df.loc[j, 'Times Cited, WoS Core'])
    else:
        CP += 1
        CPP += int(country_df.loc[j, 'Times Cited, WoS Core'])
            
df_country.loc[i, 'IP'] = IP
df_country.loc[i, 'IP%'] = IP/254654
df_country.loc[i, 'ICPP'] = IPP/IP
df_country.loc[i, 'CP'] = CP
df_country.loc[i, 'CP%'] = FP/99005
df_country.loc[i, 'CCPP'] = CPP/CP
df_country.loc[i, 'FP'] = FP
df_country.loc[i, 'FP%'] = FP/df_country.loc[i, 'P']
```

```python
data_dir = 'D:/abstract/wos_0531/bib_360446/country_number/top20/'
for i in range(20):
    country_dir = data_dir + df_country.loc[i, 'dir']
    country_df = pd.read_csv(country_dir, engine='python')
    countries = []
    for j in range(len(country_df)):
        country = country_df.loc[j, 'Countries']
        if type(country)==str:
            countries = countries + country.split(',')
    dic = {}
    for key in countries:
        dic[key]=dic.get(key,0)+1
    dic_ = sorted(dic.items(),key = lambda x:x[1],reverse = True)
    df_country.loc[i, 'MC'] = dic_[1][0]
    df_country.loc[i, 'MCP'] = dic_[1][1]
    df_country.loc[i, 'MCP%'] = dic_[1][1]/df_country.loc[i, 'P']
```
