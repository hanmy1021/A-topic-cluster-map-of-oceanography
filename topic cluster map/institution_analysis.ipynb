```python
data_dir = 'D:/abstract/wos_0531/bib_360446/drop_duplicates_353659.csv'
P = {}
C = {}
IP = {}
CIP = {}
CP = {}
CCP = {}
FP = {}
institution_df = pd.read_csv(data_dir, engine='python')
for j in range(len(institution_df)):
    if type(institution_df.loc[j, 'Addresses']) != float:
        country_address = re.sub('\[.*?\]','',institution_df.loc[j, 'Addresses'])
        addresses = country_address.split(';')
        institutions = []
        for a in addresses:
            institution_name = a.split(',')[0].strip()
            institutions.append(institution_name)
            if a == addresses[0]:
                FP[institution_name]=FP.get(institution_name,0)+1
        institutions = list(set(institutions))
        if len(institutions) == 1:
            IP[institution_name]=IP.get(institution_name,0)+1
            CIP[institution_name]=CIP.get(institution_name,0)+institution_df.loc[j, 'Times Cited, WoS Core']
            P[institution_name]=P.get(institution_name,0)+1
            C[institution_name]=C.get(institution_name,0)+institution_df.loc[j, 'Times Cited, WoS Core']
        else:
            for ins in institutions:
                CP[ins]=CP.get(ins,0)+1
                CCP[ins]=CCP.get(ins,0)+institution_df.loc[j, 'Times Cited, WoS Core']
                P[ins]=P.get(ins,0)+1
                C[ins]=C.get(ins,0)+institution_df.loc[j, 'Times Cited, WoS Core']
```
