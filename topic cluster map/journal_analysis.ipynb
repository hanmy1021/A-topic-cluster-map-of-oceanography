```python
import pandas as pd
import os
import re
import numpy as np
```

```python
data_dir = 'D:/abstract/wos_0531/bib_360446/drop_duplicates_353659.csv'
journal_df = pd.read_csv(data_dir, engine='python')
```

```python
P = {}
C = {}
for i in range(len(journal_df)):
    journal = journal_df.loc[i, 'Source Title'].strip()
    P[journal]=P.get(journal,0)+1
    citation = journal_df.loc[i, 'Times Cited, WoS Core']
    if np.isnan(citation):
        C[journal]=C.get(journal,0)+0
    else:
        C[journal]=C.get(journal,0)+citation
```

```python
P_s = sorted(P.items(),key = lambda x:x[1],reverse = True)
```

```python
P_s
```

```python
sorted(C.items(),key = lambda x:x[1],reverse = True)
```

```python
jour_df = pd.DataFrame()
```

```python
for i in range(20):
    jour_df.loc[i, 'Journal'] = P_s[i][0]
    jour_df.loc[i, 'P'] = P_s[i][1]
    jour_df.loc[i, 'P%'] = P_s[i][1]/353659
    jour_df.loc[i, 'C'] = C[P_s[i][0]]
    jour_df.loc[i, 'CPP'] = C[P_s[i][0]]/P_s[i][1]
```
