```python
import pandas as pd
import os
import re
import numpy as np
```

```python
data_dir = 'D:/abstract/wos_0531/bib_360446/drop_duplicates_353659.csv'
df = pd.read_csv(data_dir, engine='python')
```

```python
df['WoS Categories']
```

```python
list(df['WoS Categories'])
```

```python
wos_categories = []
for category in list(df['WoS Categories']):
    if type(category)!=float:
        cs = category.split(';')
        for c in cs:
            wos_categories.append(c.strip())
```

```python
wos_categories
```

```python
wos_dic = {}
for key in wos_categories:
    wos_dic[key]=wos_dic.get(key,0)+1
```

```python
sorted(wos_dic.items(),key = lambda x:x[1],reverse = True)
```
