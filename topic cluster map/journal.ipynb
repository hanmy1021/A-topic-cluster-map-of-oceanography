```python
import pandas as pd
```

```python
df = pd.read_csv('D:/abstract/wos_0531/abstract_30y.csv', engine='python')
```

```python
journal = list(df['Source Title'])
```

```python
journal
```

```python
dic={}
for key in journal:
    dic[key]=dic.get(key,0)+1
```

```python
dic
```

```python
dic_ = sorted(dic.items(),key = lambda x:x[1],reverse = True)
```

```python
df_journal = pd.DataFrame(dic_)
```

```python
df_journal
```

```python
df_journal.columns = [['Journal', 'Publication number']]
```

```python
df_journal.to_csv('D:/abstract/wos_0531/bib_360446/journal_number.csv', index=False)
```
