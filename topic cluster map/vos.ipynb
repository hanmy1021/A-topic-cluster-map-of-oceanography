```python
vos = 'FN Clarivate Analytics Web of Science\nVR 1.0\n'
for i in range(100):
    PT = df.loc[i, 'Publication Type']
    #print('PT:', PT)
    AU = df.loc[i, 'Authors']
    #print('AU:', AU)
    TI = df.loc[i, 'Article Title']
    #print('TI:', TI)
    SO = df.loc[i, 'Source Title']
    #print('SO:', SO)
    AB = df.loc[i, 'Abstract']
    #print('AB:', AB)
    OI = df.loc[i, 'ORCIDs']
    #print('OI:', OI)
    TC = df.loc[i, 'Times Cited, WoS Core']
    #print('TC:', TC)
    Z9 = df.loc[i, 'Times Cited, All Databases']
    #print('Z9:', Z9)
    SN = df.loc[i, 'ISSN']
    #print('SN:', SN)
    EI = df.loc[i, 'eISSN']
    #print('EI:', EI)
    PD = df.loc[i, 'Publication Date']
    #print('PD:', PD)
    PY = df.loc[i, 'Publication Year']
    #print('PY:', PY)
    VL = df.loc[i, 'Volume']
    #print('VL:', VL)
    IS = df.loc[i, 'Issue']
    #print('IS:', IS)
    BP = df.loc[i, 'Start Page']
    #print('BP:', BP)
    EP = df.loc[i, 'End Page']
    #print('EP:', EP)
    AR = df.loc[i, 'Article Number']
    #print('AR:', AR)
    DI = df.loc[i, 'DOI']
    #print('DI:', DI)
    UT = df.loc[i, 'UT (Unique ID)']
    #print('UT:', UT)
    
    if type(PT) != float:
        vos = vos + 'PT ' + str(PT) + '\n'
    elif not np.isnan(PT):
        vos = vos + 'PT ' + str(PT) + '\n'

    AU_list = AU.split(';')    
    for au in AU_list:
        if au == AU_list[0]:
            vos = vos + 'AU ' + str(au) + '\n'
        else:
            vos = vos + '   ' + str(au) + '\n'

    if type(TI) != float:
        vos = vos + 'TI ' + str(TI) + '\n'
    elif not np.isnan(TI):
        vos = vos + 'TI ' + str(TI) + '\n'

    if type(SO) != float:
        vos = vos + 'SO ' + str(SO) + '\n'
    elif not np.isnan(SO):
        vos = vos + 'SO ' + str(SO) + '\n'

    if type(AB) != float:
        vos = vos + 'AB ' + str(AB) + '\n'
    elif not np.isnan(AB):
        vos = vos + 'AB ' + str(AB) + '\n'

    if type(OI) != float:
        vos = vos + 'OI ' + str(OI) + '\n'
    elif not np.isnan(OI):
        vos = vos + 'OI ' + str(OI) + '\n'

    if type(TC) != float:
        vos = vos + 'TC ' + str(TC) + '\n'
    elif not np.isnan(TC):
        vos = vos + 'TC ' + str(TC) + '\n'

    if type(Z9) != float:
        vos = vos + 'Z9 ' + str(Z9) + '\n'
    elif not np.isnan(Z9):
        vos = vos + 'Z9 ' + str(Z9) + '\n'

    if type(SN) == str:
        vos = vos + 'SN ' + str(SN) + '\n'
    elif not np.isnan(SN):
        vos = vos + 'SN ' + str(SN) + '\n'

    if type(EI) != float:
        vos = vos + 'EI ' + str(EI) + '\n'
    elif not np.isnan(EI):
        vos = vos + 'EI ' + str(EI) + '\n'

    if type(PD) != float:
        vos = vos + 'PD ' + str(PD) + '\n'
    elif not np.isnan(PD):
        vos = vos + 'PD ' + str(PD) + '\n'

    if type(PY) != float:
        vos = vos + 'PY ' + str(PY) + '\n'
    elif not np.isnan(PY):
        vos = vos + 'PY ' + str(PY) + '\n'

    if type(VL) != float:
        vos = vos + 'VL ' + str(VL) + '\n'
    elif not np.isnan(VL):
        vos = vos + 'VL ' + str(VL) + '\n'

    if type(IS) != float:
        vos = vos + 'IS ' + str(IS) + '\n'
    elif not np.isnan(IS):
        vos = vos + 'IS ' + str(IS) + '\n'

    if type(BP) != float:
        vos = vos + 'BP ' + str(BP) + '\n'
    elif not np.isnan(BP):
        vos = vos + 'BP ' + str(BP) + '\n'

    if type(EP) != float:
        vos = vos + 'EP ' + str(EP) + '\n'
    elif not np.isnan(EP):
        vos = vos + 'EP ' + str(EP) + '\n'

    if type(AR) != float:
        vos = vos + 'AR ' + str(AR) + '\n'
    elif not np.isnan(AR):
        vos = vos + 'AR ' + str(AR) + '\n'

    if type(DI) != float:
        vos = vos + 'DI ' + str(DI) + '\n'
    elif not np.isnan(DI):
        vos = vos + 'DI ' + str(DI) + '\n'

    if type(UT) != float:
        vos = vos + 'UT ' + str(UT) + '\n'
    elif not np.isnan(UT):
        vos = vos + 'UT ' + str(UT) + '\n'

    vos = vos + 'ER\n\n'
vos = vos + 'EF'
```
