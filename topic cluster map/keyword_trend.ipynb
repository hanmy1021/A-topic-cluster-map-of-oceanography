```python
data_dir = 'D:/abstract/wos_0531/bib_360446/drop_duplicates_353659.csv'
df = pd.read_csv(data_dir, engine='python')
```

```python
l = []
for i in list(df['Author Keywords']):
    if type(i)!=float:
        ls = i.split(';')
        for m in ls:
            if m.strip():
                m = m.strip().upper()
                l.append(m.strip())
    
```

```python
dic = {}
for key in l:
    dic[key]=dic.get(key,0)+1
```

```python
sorted(dic.items(),key = lambda x:x[1],reverse = True)
```

```python
keywords = ['CLIMATE CHANGE','REMOTE SENSING','PHYTOPLANKTON','PRECIPITATION','SEDIMENT','SEA ICE','SEDIMENT TRANSPORT','TEMPERATURE',
           'STABLE ISOTOPES','WATER QUALITY','ENSO','DROUGHT','NUTRIENTS','ARCTIC','SOIL MOISTURE','SEA SURFACE TEMPERATURE','HOLOCENE',
           'EVAPOTRANSPIRATION','GROUNDWATER','MEDITERRANEAN SEA','SALINITY','GIS','ZOOPLANKTON','SOUTH CHINA SEA','BALTIC SEA','GROWTH',
           'ESTUARY','UPWELLING','HYDROLOGY','MODIS']
years = [i for i in range(1992,2022)]
arr = np.zeros((30,30))
df_key = pd.DataFrame(arr, columns=[keywords], index=[years], dtype=int)
```

```python
df_key
```

```python
df['Author Keywords Upper'] = df['Author Keywords']
```

```python
keyword_upper = list(df['Author Keywords Upper'])
```

```python
keyword_upper
```

```python
for i in range(len(keyword_upper)):
    if type(keyword_upper[i]) != float:
        df.loc[i, 'Author Keywords Upper'] = keyword_upper[i].upper()
```

```python
for key in keywords:
    for i in range(len(df)):
        if type(df.loc[i, 'Author Keywords Upper'])!=float and key in df.loc[i, 'Author Keywords Upper']:
            if df.loc[i, 'Publication Year'] and df.loc[i, 'Publication Year']<2022:
                df_key.loc[df.loc[i, 'Publication Year'], key] += 1
```
